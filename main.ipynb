```python
import math
from random import randint
import numpy as np
```

```python
class Kmeans:
    def __init__(self, data, k):
        self.data = data
        self.k = k

    def cluster(self) -> dict():
        clusters = {}
        for i in range(self.k):
            clusters[i] = []

        # Select k unique random centroids
        s = [i for i in range(len(self.data))]
        for i in range(self.k):
            r = randint(0, len(s)-1)
            clusters[i].append(self.data[s.pop(r)])

        centroids = [clusters[x][0] for x in clusters]
        # Put each point in the cluster with the nearest centroid
        for p in self.data:
            if p not in centroids:
                cluster_i = 0
                shortest = self.eucl(clusters[0][0], p)
                for i in range(self.k):
                    d = self.eucl(clusters[i][0], p)
                    if d < shortest:
                        shortest = d
                        cluster_i = i
                clusters[cluster_i].append(p)

        return clusters

    def eucl(self, x, y) -> float:
        return math.sqrt((x - y)**2)

    def mean(self, cluster) -> float:
        return sum(cluster) / len(cluster)
```

```python
#data = np.random.uniform(0, 20, 10)
data = [1, 2, 3, 4, 9, 10, 12, 34, 38, 40]
kmeans = Kmeans(data, 3)
clusters = kmeans.cluster()
for i in clusters:
    print(clusters[i])
```
